```python
# Definimos una función para parsear nuestro conjunto de datos
import librosa
import numpy as np
def parse_dataset(dataset):
    X = []
    y = []
    try:
        for xs in range(len(dataset)):
            audio_value = dataset[xs]['audio']
            D = librosa.amplitude_to_db(np.abs(librosa.stft(audio_value)), ref=np.float32)
            X.append(D)
            X_train_prep = np.array(X).reshape((len(X), 1025*44))
           
            encoded_value = dataset[xs]['species_encoded']
            y.append(encoded_value)
    except:
        print(xs)
        audio_value = dataset[xs]['audio_name']
    return (X, y)
```

```python
X, y = parse_dataset(dataset)
```

```python
print("La longitud del conjunto de datos es: ", len(X))
```
